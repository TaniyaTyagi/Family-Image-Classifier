```python
import pandas as pd
import numpy as np
import cv2
import os
from PIL import Image
import warnings
warnings.filterwarnings("ignore")
```

```python
# Step 1: List of all images

image_list = [
    "Images\Chote papa ji 1.jpg","Images\Chote papa ji 2.jpg","Images\Chote papa ji 3.jpg","Images\Chote pap ji 4.jpg","Images\Chote papa ji 5.jpg",
    "Images\Lakshay 1.jpg","Images\Lakshay 2.jpg","Images\Lakshay 3.jpg","Images\Lakshay 4.jpg","Images\Lakshay 5.jpg",
    "Images\Mummy 1.jpg","Images\Mummy 2.jpg","Images\Mummy 3.jpg","Images\Mummy 4.jpg","Images\Mummy 5.jpg",
    "Images\Pita ji 1.jpg","Images\Pita ji 2.jpg","Images\Pita ji 3.jpg","Images\Pita ji 4.jpg","Images\Pita ji 5.jpg",
    "Images\Taniya 1.jpg","Images\Taniya 2.jpg","Images\Taniya 3.jpg","Images\Taniya 4.jpg","Images\Taniya 5.jpg"
]
```

```python
# Step 2: Convert images into 1D arrays and stack into 

def create_2d_array_from_images(image_list, target_size=(128, 128)):
    all_1d_arrays = []
    for image_path in image_list:
        try:
            # 1. Open image
            img = Image.open(image_path).convert("RGB")
            # 2. Resize
            img_resized = img.resize(target_size)
            # 3. Convert to numpy
            img_array = np.array(img_resized)
            # 4. Flatten to 1D
            flattened_array = img_array.flatten()
            # Add to list
            all_1d_arrays.append(flattened_array)
        except Exception as e:
            print(f"Skipping image {image_path} due to error: {e}")

    if all_1d_arrays:
        return np.array(all_1d_arrays)
    else:
        return None

final_2d_array = create_2d_array_from_images(image_list, target_size=(128,128))
print("Shape of final data array (X):", final_2d_array.shape)
```

```python
# Step 3: Labels (numeric only)

# Example: 0=Chote papa, 1=Lakshay, 2=Mummy, 3=Pita ji, 4=Taniya
image_labels = [
    0,0,0,0,0,   # Chote papa’s 5 images
    1,1,1,1,     # Lakshay’s 5 images
    2,2,2,2,2,   # Mummy’s 5 images
    3,3,3,3,3,   # Pita ji’s 5 images
    4,4,4,4,4    # Taniya’s 5 images
]

labels_np = np.array(image_labels)
labels_2d = labels_np.reshape(-1, 1)

print("Shape of labels array (y):", labels_2d.shape)
```

```python
# Step 4: Train-Test Split

from sklearn.model_selection import train_test_split

x = final_2d_array
y = labels_np   # use 1D (important for sklearn)

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2)
print(train_x.shape, train_y.shape)
```

```python
# Step 5: Train Logistic Regression

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(train_x, train_y)
print("Model training completed!")
```

```python
# Step 6: Save model with pickle

import pickle

pickle.dump(model, open("family_model.pkl", "wb"))
print(" Model saved as family_model.pkl")
```
